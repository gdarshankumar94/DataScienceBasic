```python
#Assigning Target Variable & 
X = car_eval.drop(['class'], axis=1)

y = car_eval['class']
```

```python
print(X)
print(y)
```

```python
# Spliting entire dataset into Training and Test Dataset
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.5, random_state = 55)
```

```python
#Check Dataset and Trainset

X_train.shape, X_test.shape
```

```python
#converting ordinal data to categorical datatype



X_train = pd.get_dummies(X_train,columns=['buying', 'maint', 'doors', 'persons', 'luggage', 'safety'])

X_test = pd.get_dummies(X_test,columns=['buying', 'maint', 'doors', 'persons', 'luggage', 'safety'])
```

```python
#Calling Decision Tree Classifier
# import DecisionTreeClassifier

from sklearn.tree import DecisionTreeClassifier
```

```python
# Classification based on Entropy

from sklearn.tree import DecisionTreeClassifier

entropy_std = DecisionTreeClassifier(criterion='entropy', random_state=0)
```

```python
# Fit the above Model to Training Dataset

entropy_std.fit(X_train, y_train)
```

```python
#Predicting values of Y based on Entropy X_test Dataset

en_pre_y = entropy_std.predict(X_test)
```

```python
#Getting accuracy

from sklearn.metrics import accuracy_score

ac = accuracy_score(y_test,en_pre_y)

print(ac)
```

```python
# Get Accuracy of Trainset over Predicted

entropy_pr_y_train = entropy_std.predict(X_train)
```

```python
#Getting Accuracy

from sklearn.metrics import accuracy_score

acc_train = accuracy_score(y_train, entropy_pr_y_train)

print(acc_train)
```
